##

 ## Step 1 - Ask the user for two numbers

```python
n1= int(input("Enter your first number:"))
n2= int(input("Enter your second numberr:"))
```

## Step 2- Ask which operation

```python
operation = input("Choose +  -  *  /  %  // : ")
```

## Step 3 - Doing Calculations

```python
if operation == "+":
    print("Answer:", n1 + n2)

elif operation == "-":
    print("Answer:", n1 - n2)

elif operation == "*":
    print("Answer:", n1 * n2)

elif operation == "/":
    print("Answer:", n1 / n2)

elif operation == "%":
    print("Answer:", n1 % n2) # remainder

elif operation == "//":
    print("Answer:", n1 // n2) # Floor division

else:
    print("Invalid operation")
```
